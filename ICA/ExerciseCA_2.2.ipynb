```python
import matplotlib.pyplot as plt
import numpy as np
N = 1000000
x = np.random.rand(N)
y = np.random.rand(N)

# The inverse of the Laplacian CDF
def inv_laplace(x):
    return (1/lmd)*(-0.5)*np.sign(x-0.5)*np.log(1-2*np.abs(x-0.5))

lmd = np.sqrt(2)/2

# Here we turn uniformly distributed samples into Laplace-distributed samples; see p139 of Papoulis
gx = inv_laplace(x)
gy = inv_laplace(y)

g = np.array([gx, gy]).T

plt.plot(g[:,0], g[:,1], '.', alpha=0.4)

m = np.abs(g).max()
plt.gca().set_aspect('equal')
plt.xlim((-m,m))
plt.ylim((-m,m));
```

```python
# Here we turn uncorrelated Laplacian RVs into correlated RVs (target cov matrix: C)

C = np.array([[4, -1],[-1, 2]])
R = np.linalg.cholesky(C)

R@R.T
g.T@g/N

h = (R@g.T).T


plt.plot(h[:,0], h[:,1], '.', alpha=0.4)

m = np.abs(h).max()
plt.gca().set_aspect('equal')
plt.xlim((-m,m))
plt.ylim((-m,m))

h.T@h/N
#plt.hist(Gx, 100)
```
