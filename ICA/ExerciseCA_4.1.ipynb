```python
df = pd.read_csv('./data/automobile/imports-85.data')

d = np.array(df)
# 16, 21, 23, 24
# 24 is nonlinear

x =  d[:,24]
y = d[:,25]

delix1 = np.where(np.array(x) == '?')[0]
delix2 = np.where(np.array(y) == '?')[0]
delix = np.concatenate((delix1, delix2), axis=0)

x = np.delete(x, delix)
y = np.delete(y, delix)
x = x.astype(float)
y = y.astype(float)

H = np.concatenate([[np.array(x)], [np.ones(x.shape)]]).T
HTH = H.T@H
Hmpi = np.linalg.inv(HTH)@H.T
z  = Hmpi @ y
print(z)
plt.plot(x, y, 'x')
```
